### Objetivo Específico 1:
Identificar la categoría de productos más vendidos (monto de dinero y cantidad) en su evolución mensual.




```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as tick
import seaborn as sns
```

```python
top_categorias_por_cantidad_ventas = pd.read_sql_query(
    '''
    SELECT product_category_name,COUNT(oi.product_id) as cantidad
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    GROUP BY product_category_name
    ORDER BY cantidad DESC
    LIMIT 10
    ''',
    conn
)

# products tiene product_id y product_category_name
# order_items tiene product_id

top_categorias_por_cantidad_ventas
```

```python
bars = plt.bar(top_categorias_por_cantidad_ventas['product_category_name'], top_categorias_por_cantidad_ventas['cantidad'])

plt.xlabel('Categorías de Productos')
plt.ylabel('Cantidad de Ventas')
plt.title("Cantidad de Ventas según Categorías de Productos")
plt.tick_params(axis='x', labelrotation=90)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.savefig('categoria_productos_mas_vendidas.png')
```

En el presente gráfico podemos observar el **Top 10 de Categorías de Productos** más populares en términos de **cantidad de ventas**. En este sentido, la categoría **"cama_mesa_banho"** ocupa el **primer lugar** con poco más de **11.000 productos vendidas**, seguido de **"beleza_saude"** con poco **menos de 10.000**. Dentro de este Top 10, el último de los peldaños lo ocupa "automotivo" con más de 4.000 productos vendidos de esta categoría. Aquí podemos ver, por lo tanto, **los productos de mayor interés para nuestros clientes**, lo cual nos puede servir con un análisis más profundo para calcular el stock necesario, definir una estrategia adecuada de pricing y marketing.

```python
top_categoria_por_monto_operado = pd.read_sql_query(
    '''
    SELECT product_category_name,COUNT(oi.product_id) as cantidad, SUM(payment_value) as monto_operado
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    GROUP BY product_category_name
    ORDER BY monto_operado DESC
    LIMIT 10
    ''',
    conn
)


top_categoria_por_monto_operado
```

```python
## La función reformat_large_tick_values() sirve para achicar los números con notación científica o demasiado largo para el gráfico

def reformat_large_tick_values(tick_val, pos):
    if tick_val >= 1000000000:
        val = round(tick_val/1000000000, 1)
        new_tick_format = '{:}B'.format(val)
    elif tick_val >= 1000000:
        val = round(tick_val/1000000, 1)
        new_tick_format = '{:}M'.format(val)
    elif tick_val >= 1000:
        val = round(tick_val/1000, 1)
        new_tick_format = '{:}K'.format(val)
    elif tick_val < 1000:
        new_tick_format = round(tick_val, 1)
    else:
        new_tick_format = tick_val

    new_tick_format = str(new_tick_format)

    index_of_decimal = new_tick_format.find(".")

    if index_of_decimal != -1:
        value_after_decimal = new_tick_format[index_of_decimal+1]
        if value_after_decimal == "0":
            new_tick_format = new_tick_format[0:index_of_decimal] + new_tick_format[index_of_decimal+2:]

    return new_tick_format

fig, ax = plt.subplots()
ax.yaxis.set_major_formatter(tick.FuncFormatter(reformat_large_tick_values))

bars = plt.bar(top_categoria_por_monto_operado['product_category_name'], top_categoria_por_monto_operado['monto_operado'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, reformat_large_tick_values(yval, None),
             ha='center', va='bottom', fontsize=8)

plt.xlabel('Categorías de Productos')
plt.ylabel('Dinero Operado en Millones de Reales')
plt.title("Cantidad de Ventas según Categorías de Productos")
plt.tick_params(axis='x', labelrotation=90)

plt.savefig('categoria_productos_monto_operado.png')
```

En el gráfico de arriba podemos observar el TOP 10 de las categorías de productos con mayor monto operado, en millones de reales. Podemos ver que  "cama_mesa_banho" y "beleza_saude" mantienen el top 1 y 2 respectivamente, al igual que cuando las medimos por cantidad de productos. Sin embargo, el puesto 3 pasa a ocuparlo informatica_accesorias en el caso de Monto Operado, cuando esta categoría detentaba el puesto 5 en cantidad de productos vendidos, indicando que el precio promedio de esta categoría es probablemente en promedio más alta frente a otras categorías de productos. Otro punto a destacar es que "telefonia" ya no aparece en nuestro gráfico de Monto, y aparece una nueva categoría en la última posición del TOP 10, siendo esta "cool_stuff"

#### Análisis de evolución mensual de las categorías con mayor cantidad de compras

```python
## Creamos una tabla temporal con el TOP 10 de categorías de productos más vendidos

top_10_categorias_por_cantidad = '''
CREATE TEMPORARY TABLE top_10_categorias_por_cantidad AS
    SELECT product_category_name,COUNT(oi.product_id) as cantidad
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    GROUP BY product_category_name
    ORDER BY cantidad DESC
    LIMIT 10;
'''


conn.execute(top_10_categorias_por_cantidad)


top_10_categorias_por_cantidad = pd.read_sql_query('SELECT * FROM top_10_categorias_por_cantidad', conn)


print(top_10_categorias_por_cantidad)
```

```python
casa_mesa_banho_mensual = pd.read_sql_query(
    '''
    SELECT  product_category_name,
            strftime('%Y', o.order_purchase_timestamp) AS Year,
            strftime('%m', o.order_purchase_timestamp) AS Month,
            COUNT(oi.product_id) as cantidad
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    INNER JOIN orders o ON
    op.order_id=o.order_id
    INNER JOIN customers c ON
    c.customer_id=o.customer_id
    WHERE product_category_name != "None"
    GROUP BY product_category_name,year,month
    HAVING product_category_name IN ( SELECT product_category_name
                                      FROM top_10_categorias_por_cantidad
                                      ORDER BY cantidad DESC
                                      LIMIT 10)
    ORDER BY product_category_name,year,month
    ''',
    conn
)


casa_mesa_banho_mensual
```

```python
# Crear una columna de fecha única
casa_mesa_banho_mensual['date'] = casa_mesa_banho_mensual['Year'].astype(str) + '-' + casa_mesa_banho_mensual['Month'].astype(str).str.zfill(2)
casa_mesa_banho_mensual['date'] = pd.to_datetime(casa_mesa_banho_mensual['date'], format='%Y-%m')

# Calcular el promedio de cantidad de productos vendidos para todas las categorías
promedio = casa_mesa_banho_mensual['cantidad'].mean()

# Configurar la figura y los ejes
plt.figure(figsize=(15, 8))

# Dibujar una línea horizontal para el promedio
plt.axhline(y=promedio, color='black', linestyle='--', label='Promedio',alpha=0.3)

# Anotar el valor del promedio en la gráfica
plt.text(casa_mesa_banho_mensual['date'].iloc[-1], promedio, f'${promedio:.2f}', ha='left', va='center', color='black',
         bbox=dict(boxstyle='round', facecolor='white', edgecolor='black'))

sns.lineplot(x='date', y='cantidad', hue='product_category_name', data=casa_mesa_banho_mensual, marker='o')

plt.title('Top 10 Categorías de Producto por Cantidad de Ventas')
plt.xlabel('Fecha (Año-Mes)')
plt.ylabel('$')

# Remover la notación científica del eje Y
plt.ticklabel_format(style='plain', axis='y')
```

En principio, es posible observar como el flujo operado a mediados de 2016 es notablemente bajo, posiblemente correlacionado con el inicio de la operación de la compañía. Asimismo, vemos un incremento constante y un pico fuerte en cantidad de ventas hacia finales de 2017, que luego tiende a estabilizarse. A su vez, la categoría "informatica_acessorios" tiene un pico más tardío en el primer trimestre de 2018 y una caída más pronunciada en cantidad de ventas que otras categorías. Sin embargo, viendo tantas categorías a la vez, puede ser un poco confuso, con lo cual, realizaremos nuevamente la query y gráfico sólo con el top 3 de categorías más vendidas.

```python
casa_mesa_banho_mensual_top3 = pd.read_sql_query(
    '''
    SELECT  product_category_name,
            strftime('%Y', o.order_purchase_timestamp) AS Year,
            strftime('%m', o.order_purchase_timestamp) AS Month,
            COUNT(oi.product_id) as cantidad
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    INNER JOIN orders o ON
    op.order_id=o.order_id
    INNER JOIN customers c ON
    c.customer_id=o.customer_id
    WHERE product_category_name != "None"
    GROUP BY product_category_name,year,month
    HAVING product_category_name IN ( SELECT product_category_name
                                      FROM top_10_categorias_por_cantidad
                                      ORDER BY cantidad DESC
                                      LIMIT 3)
    ORDER BY product_category_name,year,month
    ''',
    conn
)


casa_mesa_banho_mensual_top3
```

```python
# Crear una columna de fecha única
casa_mesa_banho_mensual_top3['date'] = casa_mesa_banho_mensual_top3['Year'].astype(str) + '-' + casa_mesa_banho_mensual_top3['Month'].astype(str).str.zfill(2)
casa_mesa_banho_mensual_top3['date'] = pd.to_datetime(casa_mesa_banho_mensual_top3['date'], format='%Y-%m')

# Calcular el promedio de cantidad de productos vendidos para todas las categorías
promedio = casa_mesa_banho_mensual_top3['cantidad'].mean()

# Configurar la figura y los ejes
plt.figure(figsize=(15, 8))

# Dibujar una línea horizontal para el promedio
plt.axhline(y=promedio, color='black', linestyle='--', label='Promedio',alpha=0.3)

# Anotar el valor del promedio en la gráfica
plt.text(casa_mesa_banho_mensual_top3['date'].iloc[-1], promedio, f'${promedio:.2f}', ha='left', va='center', color='black',
         bbox=dict(boxstyle='round', facecolor='white', edgecolor='black'))

sns.lineplot(x='date', y='cantidad', hue='product_category_name', data=casa_mesa_banho_mensual_top3, marker='o')

plt.title('Top 3 Categorías de Producto por Cantidad de Ventas')
plt.xlabel('Fecha (Año-Mes)')
plt.ylabel('$')
```

Aquí ya es más entendible la información, viendo solamente a las 3 categorías con productos más vendidos. La evolución mensual es similar a la anteriormente vista, pero podemos ver que el promedio de ventas es notablemente mayor. Por otra parte, también es posible poner el foco en la categoría "esporte_lazer", dado que del segundo trimestre del 2018 en adelante, parece tener una baja más abrupta en cantidad de ventas.

### Objetivo Específico 2:
Analizar la evolución mensual de las ciudades más populares en términos de cantidad de órdenes de compra y monto operado.

```python
## Creamos una tabla temporal con el TOP 10 de ciudades por cantidad de ordenes de compra

top_10_ciudades_cantidad_ordenes = '''
CREATE TEMPORARY TABLE top_10_ciudades_cantidad_ordenes AS
    SELECT  customer_city,
            COUNT(DISTINCT oi.order_id) as cantidad_ordenes
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    INNER JOIN orders o ON
    op.order_id=o.order_id
    INNER JOIN customers c ON
    c.customer_id=o.customer_id
    GROUP BY customer_city

    ORDER BY cantidad_ordenes DESC
    LIMIT 10;
'''


conn.execute(top_10_ciudades_cantidad_ordenes)


top_10_ciudades_cantidad_ordenes = pd.read_sql_query('SELECT * FROM top_10_ciudades_cantidad_ordenes', conn)


print(top_10_ciudades_cantidad_ordenes)
```

```python
evolucion_cantidad_ordenes_ciudad = pd.read_sql_query(
    '''
    SELECT  customer_city,
            strftime('%Y', o.order_purchase_timestamp) AS Year,
            strftime('%m', o.order_purchase_timestamp) AS Month,
            COUNT(DISTINCT oi.order_id) as cantidad_ordenes
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    INNER JOIN orders o ON
    op.order_id=o.order_id
    INNER JOIN customers c ON
    c.customer_id=o.customer_id
    WHERE order_purchase_timestamp < '2018-09-01'  -- excluimos este mes porque solo tiene una compra y distorciona el gráfico posterior
    GROUP BY customer_city,year,month
    HAVING customer_city IN ( SELECT customer_city
                              FROM top_10_ciudades_cantidad_ordenes
                              ORDER BY cantidad_ordenes DESC)
    ORDER BY customer_city,year,month
    ''',
    conn
)


evolucion_cantidad_ordenes_ciudad
```

```python
# Crear una columna de fecha única
evolucion_cantidad_ordenes_ciudad['date'] = evolucion_cantidad_ordenes_ciudad['Year'].astype(str) + '-' + evolucion_cantidad_ordenes_ciudad['Month'].astype(str).str.zfill(2)
evolucion_cantidad_ordenes_ciudad['date'] = pd.to_datetime(evolucion_cantidad_ordenes_ciudad['date'], format='%Y-%m')

# Calcular el promedio de cantidad de órdenes de compra
promedio = evolucion_cantidad_ordenes_ciudad['cantidad_ordenes'].mean()

# Configurar la figura y los ejes
plt.figure(figsize=(15, 8))

# Dibujar una línea horizontal para el promedio
plt.axhline(y=promedio, color='black', linestyle='--', label='Promedio',alpha=0.3)

# Anotar el valor del promedio en la gráfica
plt.text(evolucion_cantidad_ordenes_ciudad['date'].iloc[-1], promedio, f'{promedio:.2f}', ha='left', va='center', color='black',
         bbox=dict(boxstyle='round', facecolor='white', edgecolor='black'))

sns.lineplot(x='date', y='cantidad_ordenes', hue='customer_city', data=evolucion_cantidad_ordenes_ciudad, marker='o')

plt.title('Top 10 Ciudades por Cantidad de Órdenes de Compra')
plt.xlabel('Fecha (Año-Mes)')
plt.ylabel('Cantidad de Órdenes de Compra')
```

Al igual que en el objetivo específico 1, vemos un patrón temporal similar donde las compras arrancan despacio en Octubre de 2016 y aumentan progresivamente hasta su pico en octubre de 2017, aunque en este caso no estamos analizando cantidad de productos sino cantidad de órdenes de compra. Más allá de eso, vemos que la ciudad de Sao Paulo no parece obedecer una estabilización en cantidad de compras sino que sigue creciendo en el 2018. Por otra lado, vemos que hay dos ciudades que superan ampliamente el promedio de órdenes de compra por mes, siendo estas la ya mencionada Sao Paulo seguida de Rio de Janeiro. Sería interesante analizar estas dos ciudades por separado, dado que evidentemente manejan un volumen de compras bastante superior al resto de ciudades de nuestro Top 10.

```python
evolucion_cantidad_ordenes_ciudad = pd.read_sql_query(
    '''
    SELECT  customer_city,
            strftime('%Y', o.order_purchase_timestamp) AS Year,
            strftime('%m', o.order_purchase_timestamp) AS Month,
            COUNT(DISTINCT oi.order_id) as cantidad_ordenes
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    INNER JOIN orders o ON
    op.order_id=o.order_id
    INNER JOIN customers c ON
    c.customer_id=o.customer_id
    WHERE order_purchase_timestamp < '2018-09-01'  -- excluimos este mes porque solo tiene una compra y distorciona el gráfico posterior
    GROUP BY customer_city,year,month
    HAVING customer_city IN ( SELECT customer_city
                              FROM top_10_ciudades_cantidad_ordenes
                              ORDER BY cantidad_ordenes DESC
                              LIMIT 2)
    ORDER BY customer_city,year,month
    ''',
    conn
)


evolucion_cantidad_ordenes_ciudad
```

Analicemos ahora si este patrón se vuelve a repetir al analizar monto de dinero operado.

```python
## Creamos una tabla temporal con el TOP 10 de ciudades por monto de dinero operado

top_10_ciudades_monto_operado = '''
CREATE TEMPORARY TABLE top_10_ciudades_monto_operado AS
    SELECT  customer_city,
            SUM(op.payment_value) as monto_operado
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    INNER JOIN orders o ON
    op.order_id=o.order_id
    INNER JOIN customers c ON
    c.customer_id=o.customer_id
    GROUP BY customer_city

    ORDER BY monto_operado DESC
    LIMIT 10;
'''


conn.execute(top_10_ciudades_monto_operado)


top_10_ciudades_monto_operado = pd.read_sql_query('SELECT * FROM top_10_ciudades_monto_operado', conn)


print(top_10_ciudades_monto_operado)
```

```python
evolucion_monto_operado_ciudad = pd.read_sql_query(
    '''
    SELECT  customer_city,
            strftime('%Y', o.order_purchase_timestamp) AS Year,
            strftime('%m', o.order_purchase_timestamp) AS Month,
            SUM(op.payment_value) as monto_operado
    FROM order_items oi
    INNER JOIN products p ON
    p.product_id=oi.product_id
    INNER JOIN order_payments op ON
    op.order_id=oi.order_id
    INNER JOIN orders o ON
    op.order_id=o.order_id
    INNER JOIN customers c ON
    c.customer_id=o.customer_id
    WHERE order_purchase_timestamp < '2018-09-01'  -- excluimos este mes porque solo tiene una compra y distorciona el gráfico posterior
    GROUP BY customer_city,year,month
    HAVING customer_city IN ( SELECT customer_city
                              FROM top_10_ciudades_monto_operado
                              ORDER BY monto_operado DESC)
    ORDER BY customer_city,year,month
    ''',
    conn
)


evolucion_monto_operado_ciudad
```

```python
## La función reformat_large_tick_values() sirve para achicar los números con notación científica o demasiado largo para el gráfico

def reformat_large_tick_values(tick_val, pos):
    if tick_val >= 1000000000:
        val = round(tick_val/1000000000, 1)
        new_tick_format = '{:}B'.format(val)
    elif tick_val >= 1000000:
        val = round(tick_val/1000000, 1)
        new_tick_format = '{:}M'.format(val)
    elif tick_val >= 1000:
        val = round(tick_val/1000, 1)
        new_tick_format = '{:}K'.format(val)
    elif tick_val < 1000:
        new_tick_format = round(tick_val, 1)
    else:
        new_tick_format = tick_val

    new_tick_format = str(new_tick_format)

    index_of_decimal = new_tick_format.find(".")

    if index_of_decimal != -1:
        value_after_decimal = new_tick_format[index_of_decimal+1]
        if value_after_decimal == "0":
            new_tick_format = new_tick_format[0:index_of_decimal] + new_tick_format[index_of_decimal+2:]

    return new_tick_format

fig, ax = plt.subplots()
ax.yaxis.set_major_formatter(tick.FuncFormatter(reformat_large_tick_values))

bars = ax.bar(metodo_de_pago['payment_type'], metodo_de_pago['total_dinero'], color=palette_color)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, reformat_large_tick_values(yval, None),
             ha='center', va='bottom', fontsize=8)

plt.xlabel('Métodos de pagos')
plt.ylabel('Dinero Operado en Millones de Reales')
plt.title("Métodos de pagos según cantidad de dinero operado")
plt.tick_params(axis='x')

plt.savefig('metodo_de_pago_dinero.png')
```

```python
palette_color = sns.color_palette("pastel")

plt.pie(metodo_de_pago['total_dinero'][0:-1], labels=metodo_de_pago['payment_type'][0:-1], colors=palette_color, autopct='%.0f%%')

plt.title("Métodos de pagos según porcentaje de dinero operado")
plt.savefig('metodo_de_pago_dinero2.png')
plt.show()
```

Veamos qué sucede cuando analizamos **los gráficos de métodos de pago pero por cantidad de dinero operado**. Nuevamente, el gráfico de barras trabaja con los números absolutos, mientras que el gráfico de tortas lo hace con porcentajes. Vemos aquí nuevamente **un patrón similar siendo cómoda la tarjeta de crédito el método a través del cual más dinero se operó, totabilizando el 78% del dinero operado (más de 12 millones de reales)**. En **segundo lugar, el boleto tiene porcentajes casi idénticos al de cantidad de operaciones, 18%** y un monto operado de cerca de $3 millones de reales. No así **el voucher que, aunque representaba el 6% de las operaciones, aquí vemos que sólo representa el 2% del dinero total operado. Esto indica que el voucher moviliza en promedio menos dinero por operación que el resto de los métodos de pago**. Por su lado, la tarjeta de débito nuevamente representa el 1% del total, en este caso del monto de dinero operado.
